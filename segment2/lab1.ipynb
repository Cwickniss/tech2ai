```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generate synthetic data
np.random.seed(42)
n_samples = 100
weight = np.random.uniform(0, 10, n_samples)
noise = np.random.normal(0, 50, n_samples)
price = 200 + 100*weight - 40*weight**2 + 4*weight**3 + noise
```

```python
# Plot the generated data
plt.figure(figsize=(10, 6))
plt.scatter(weight, price, alpha=0.5)
plt.title('Home Appliance Price vs Weight')
plt.xlabel('Weight')
plt.ylabel('Price')
plt.show()
```

```python
X = weight.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, price, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_hat_test = model.predict(X_test)
mse = mean_squared_error(y_test, y_hat_test)
r2 = r2_score(y_test, y_hat_test)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, alpha=0.5)
plt.plot(X_test, y_hat_test, color='red', linewidth=2)
plt.title(f'Linear Regression (Degree 1)\nMSE: {mse:,.0f}, R2: {r2:.2f}')
plt.xlabel('Weight')
plt.ylabel('Price')
plt.show()
```

```python
X2 = np.column_stack((weight, weight**2))
X_train, X_test, y_train, y_test = train_test_split(X2, price, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_hat_test = model.predict(X_test)
mse = mean_squared_error(y_test, y_hat_test)
r2 = r2_score(y_test, y_hat_test)
```

```python
# Sort X_test for smooth curve plotting
X_test_sorted = X_test[X_test[:, 0].argsort()]
y_hat_test_sorted = model.predict(X_test_sorted)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_test[:, 0], y_test, alpha=0.5)
plt.plot(X_test_sorted[:, 0], y_hat_test_sorted, color='red', linewidth=2)
plt.title(f'Polynomial Regression (Degree 2)\nMSE: {mse:,.0f}, R2: {r2:.2f}')
plt.xlabel('Weight')
plt.ylabel('Price')
plt.show()
```

```python
# Step 3: Polynomial Regression (Degree 3)
X3 = np.column_stack((weight, weight**2, weight**3))
X_train, X_test, y_train, y_test = train_test_split(X3, price, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_hat_test = model.predict(X_test)
mse = mean_squared_error(y_test, y_hat_test)
r2 = r2_score(y_test, y_hat_test)
```

```python
# Sort X_test for smooth curve plotting
X_test_sorted = X_test[X_test[:, 0].argsort()]
y_hat_test_sorted = model.predict(X_test_sorted)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_test[:, 0], y_test, alpha=0.5)
plt.plot(X_test_sorted[:, 0], y_hat_test_sorted, color='red', linewidth=2)
plt.title(f'Polynomial Regression (Degree 3)\nMSE: {mse:,.0f}, R2: {r2:.2f}')
plt.xlabel('Weight')
plt.ylabel('Price')
plt.show()
```

```python
# Step 4: Overfitting Demonstration (Degree 30)
X_overfit = np.column_stack([weight**i for i in range(1, 31)])
X_train, X_test, y_train, y_test = train_test_split(X_overfit, price, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_hat_test = model.predict(X_test)
mse = mean_squared_error(y_test, y_hat_test)
r2 = r2_score(y_test, y_hat_test)
```

```python
# Sort X_test for smooth curve plotting
X_test_sorted = X_test[X_test[:, 0].argsort()]
y_hat_test_sorted = model.predict(X_test_sorted)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_test[:, 0], y_test, alpha=0.5)
plt.plot(X_test_sorted[:, 0], y_hat_test_sorted, color='red', linewidth=2)
plt.title(f'Polynomial Regression (Degree 30)\nMSE: {mse:,.0f}, R2: {r2:.2f}')
plt.xlabel('Weight')
plt.ylabel('Price')
plt.show()
```
